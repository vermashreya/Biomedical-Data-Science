```python
from keras.models import Sequential
from keras.layers import Dense
import pandas as pd
from sklearn.model_selection import train_test_split
import numpy
from sklearn.preprocessing import StandardScaler
from sklearn.metrics import confusion_matrix
from keras.callbacks import ModelCheckpoint,EarlyStopping
```

```python
# fix random seed for reproducibility
seed = 7
numpy.random.seed(seed)
```

```python
from google.colab import drive 
drive.mount('/content/drive')
```

```python
df = pd.read_csv('drive/My Drive/Colab Notebooks/7 - NN/data.csv')
```

```python
df.head()
```

```python
df.describe()
```

```python
#convert to numpy array
data = df.values
```

```python
# split into features (X) and labels (Y)
X = data[:,0:7]
Y = data[:,8]
```

```python
X= StandardScaler().fit_transform(X)
```

```python
X_train, X_test, y_train, y_test = \
        train_test_split(X, Y, test_size=0.20, random_state=1)
```

```python
X_train.shape
```

```python
# create a fully connected model, as a sequential class
model = Sequential()
```

```python
# fully connected layers are defined using Dense class. We define three layers: 2 hidden layers + output layer.
model.add(Dense(units=15, input_dim=7, kernel_initializer='uniform', activation='relu'))
model.add(Dense(units=8, kernel_initializer='uniform', activation='relu'))
model.add(Dense(units=1, kernel_initializer='uniform', activation='sigmoid'))
```

```python
# Compile model

model.compile(loss='binary_crossentropy', optimizer='adam', metrics=['accuracy'])
```

```python
# Fit and evaluate using a handy Keras function to do both
checkpoint = ModelCheckpoint('drive/My Drive/Colab Notebooks/7 - NN/model.hdf5', monitor='val_accuracy', save_best_only=True)#does not work on google drive.
earlystop = EarlyStopping(monitor='val_loss', patience=4)
model.fit(X_train, y_train, nb_epoch=150, batch_size=10, validation_split= 0.3,callbacks = [checkpoint,earlystop])
```

```python
# evaluate the model
scores = model.evaluate(X_test, y_test)
print("%s: %.2f%%" % (model.metrics_names[1], scores[1]*100))
```

```python
pred = model.predict(X_test)
pred[pred>=0.5]=1
pred[pred<0.5]=0
```

```python
confusion_matrix(y_test,pred)
```
